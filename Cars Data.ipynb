### Data Cleaning
1)Remove duplicate data

2)Fix missing data

3)Delete whole rows missing data

4)Renaming columns to better names

```python
#Here Iam going to remove the duplicates
df.drop_duplicates(inplace=True)
df.head()
```

```python
#Check if duplicates are removed
df.duplicated().sum()
```

```python
#Here delete the rows that contain no data at all
df.dropna(how='all',axis=0,inplace=True)
df.head()
```

```python
#Recheck on missing data
df.isnull().sum()
```

```python
#There is some null values in Cylinders column so Iam going to replace it with meadian of cylinders
df['Cylinders'].fillna(df['Cylinders'].median(),inplace=True)
```

```python
#Recheck on missing data
df.isnull().sum()
```

```python
#Check the shape of data
df.shape
```

```python
df.rename(columns={'Make':'Brand'},inplace=True)
df.head()
```

## Exploratory Data Analysis

In EDA I plotted 2 figures 

1) The most common Brand names in Europe

2) The most expensive type of cars



```python
#Here I made a coy of main dataframe to avoid missing with main in case of furthur usage
europe_df=df.copy()
#Here I will filter the origin to only Europe
europe_df=europe_df[europe_df['Origin']=='Europe']
europe_df.head()
```

```python
#Plotting the most common Brand names in Europe with 
europe_df['Brand'].value_counts().plot(kind='pie',figsize=(8,8));
plt.title("Most Common Brand in Europe");
plt.ylabel("");
```

```python
#Here I made a coy of main dataframe to avoid missing with main in case of furthur usage
exp_df=df.copy()
#Here I deleted the $ and (since \D matches any non-decimal digit) in Invoice then changed type to int
exp_df['Invoice'] = exp_df['Invoice'].str.replace('\D', '').astype('int64')
exp_df.head()   
```

```python
exp_df.groupby(['Type'])['Invoice'].mean().plot(kind='barh');
plt.xlabel('Price');
plt.ylabel("Type of Cars");
plt.title("Average Price of Car Types");
```
